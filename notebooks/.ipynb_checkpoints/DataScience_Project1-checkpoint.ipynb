```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df_movies=pd.read_csv("../data/filmtv_movies.csv")
df_boxoffice=pd.read_csv("../data/boxoffice.csv")
```

```python
df_movies.head(10)
```

```python
df_boxoffice.head(10)
```

## Merge Movie and Box Office Dataframe by title

```python
df3=pd.merge(df_movies, df_boxoffice, on="title")
df3.head(10)
```

## Clean Data

```python
df3['is_year_equal'] = np.where(df3['year_x']==df3['year_y'], 'yes', 'no')
```

```python
df3.isnull().sum()
```

```python
df3=df3.drop('description',1)
df3=df3.drop('notes',1)
df3=df3.drop('actors',1)
```

```python
df3.isna().any()
```

```python
df3.dropna(inplace=True)
df3.head(10)
```

```python
df4=df3[df3.is_year_equal != 'no']
df4.head(10)
```

## Most of the data occurs within the 2000s

```python
totalyear = df4.groupby(['year_y']).year_y.count()
totalyear.sort_values(ascending=False).head(11)
```

```python
movies2013 = df4[df4.year_y == 2013]
movies2013.head()
```

```python
movies2017 = df4[df4.year_y == 2017]
movies2016 = df4[df4.year_y == 2016]
movies2007 = df4[df4.year_y == 2007]
movies2006 = df4[df4.year_y == 2006]
movies2015 = df4[df4.year_y == 2015]
movies2008 = df4[df4.year_y == 2008]
movies2012 = df4[df4.year_y == 2012]
movies2004 = df4[df4.year_y == 2004]
movies2014 = df4[df4.year_y == 2014]
```

```python
frames = [movies2013, movies2016, movies2007, movies2017, movies2014, movies2012, movies2006, movies2015, movies2004, movies2008, movies2006]

result = pd.concat(frames)
result.head(10)
```

## Feature Engineering

```python
for i,row in result.iterrows():
    country=str(row['country'])
    if country!=('no genres listed'):
        country=country.split(',')
        for g in country:
            result.at[i,g]=1
result.head(10)
```

```python
result.fillna(0)
result.head()
```

## Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
sns.lmplot(x='rank',y='lifetime_gross',data=result, fit_reg=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
sns.lmplot(x='avg_vote',y='lifetime_gross',data=result, fit_reg=False, hue='genre')
```

```python
sPlot= sns.swarmplot(x='genre',y='lifetime_gross',data=result)
sPlot.set_xticklabels(sPlot.get_xticklabels(), rotation=90)
```

## Conclusions 
I merged two datasets together that had box office data and film information data. Since majority of my data occurred within last ten years, I decided to investigate the data from the 2000s. Through investigating this data, I found that rank and lifetime gross are exponentially related where first ranked yields a higher Lifetime gross for domestic US sales. I also found that moves that have a higher lifetime gross are not ranked exceptionally well by IMDB. They are usually ranked at about a 6/10. I also found that the fantasy and super hero genres have the highest lifetime gross in the 2000s. This is probably due to franchises like MARVEL and Harry Potter. I used heat maps to look for correlations between rank, lifetime gross, duration, and votes and also between which countries. I found a surprising but fairly strong correlation between lifetime gross and duration of the movie, and that the United States does not film with other countries.
