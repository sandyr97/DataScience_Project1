```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df_movies=pd.read_csv("../data/filmtv_movies.csv")
df_boxoffice=pd.read_csv("../data/boxoffice.csv")
```

## Merge Movie and Box Office Dataframe by title

```python
df3=pd.merge(df_movies, df_boxoffice, on="title")
df3
```

## Clean Data

```python
df3['is_year_equal'] = np.where(df3['year_x']==df3['year_y'], 'yes', 'no')
```

```python
df3.isnull().sum()
```

```python
df3=df3.drop('description',1)
df3=df3.drop('notes',1)
df3=df3.drop('actors',1)
```

```python
df3.isna().any()
```

```python
df3.dropna(inplace=True)
df3
```

```python
df4=df3[df3.is_year_equal != 'no']
df4
```

```python
totalyear = df4.groupby(['year_y']).year_y.count()
totalyear
```

```python
totalyear.max()
```

2013 has most movie data so we are going to investigate it

```python
movies2013 = df4[df4.year_y == 2013]
movies2013
```

## Feature Engineering

```python
for i,row in movies2013.iterrows():
    country=str(row['country'])
    if country!=('no genres listed'):
        country=country.split(',')
        for g in country:
            movies2013.at[i,g]=1
movies2013
```

```python
movies2013.fillna(0)
```

## Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="darkgrid")
sns.lmplot(x='rank',y='lifetime_gross',data=movies2013, fit_reg=False)
```
